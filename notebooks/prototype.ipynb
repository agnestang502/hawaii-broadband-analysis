```python
import osmnx as ox
import geopandas as gpd
import os
from pathlib import Path
```

```python
waianae_gdf= ox.geocode_to_gdf( "Waianae, Hawaii")
```

```python
#https://osmnx.readthedocs.io/en/stable/user-reference.html
geometry = waianae_gdf.geometry.iloc[0]

waianae_drive_graph=ox.graph_from_polygon(

    geometry,
    network_type="drive"

)    
gdf_drive_edges=ox.graph_to_gdfs(waianae_drive_graph, nodes=False
)
gdf_drive_edges.to_file("waianae_drive.geojson",driver="GeoJSON")
gdf_drive_edges.head()
```

```python
waianae_walk_graph=ox.graph_from_polygon(

    geometry,
    network_type="walk"

)    
gdf_walk_edges=ox.graph_to_gdfs(waianae_walk_graph, nodes=False
)
gdf_walk_edges.to_file("waianae_walk.geojson",driver="GeoJSON")
gdf_walk_edges.head()
```

```python
#https://docs.python.org/3/library/pathlib.html
data_folder = Path("data/raw")
data_folder.mkdir(parents=True, exist_ok=True)

path_drive = data_folder / "waianae_roads_drive.geojson"
path_walk = data_folder / "waianae_roads_walk.geojson"

gdf_drive_edges.to_file(path_drive, driver="GeoJSON")
gdf_walk_edges.to_file(path_walk, driver="GeoJSON")
```
