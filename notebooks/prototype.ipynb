```python
import osmnx as ox
import pandas as pd
```

```python
ox.settings.use_cache = True
ox.settings.log_console = True

clusters = {
    # "urban_honolulu": (21.304547, -157.855676),
    # "suburban_mililani": (21.4508308, -158.0095783),
    # "rural_waianae": (21.4454875, -158.1878239),
    "remote_Nāʻālehu": (19.0608, -155.5866)
}


for name, query in clusters.items():
    # print(f"Downloading street network for: {name}...")
    # coords = ox.geocode("Kaʻū, Hawaii, USA")
    # https://osmnx.readthedocs.io/en/stable/user-reference.html
    try:
        graph = ox.graph_from_point(query, 8000, network_type='drive')
        ox.save_graphml(graph, filepath=f"../data/raw/{name}_network.graphml")
        print(f"Successfully saved {name} network.")
    except Exception as e:
        print(f"Error downloading {name}: {e}")
```
